```python
#concatenating the dataframe
india_weather =pd.DataFrame({'city':['Delhi','Ujjain','Banalore'], 'temperature':[32,45,67],'humidity': [45,46,47]})
india_weather
```

```python
india_weather
```

```python
us_weather =pd.DataFrame({'city':['chicago','orlando','new_york'],'temperature':[23,54,32],'humidity':[21,33,43]})
us_weather
```

```python
df =pd.concat([india_weather,us_weather])
df
```

```python
df =pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
df = pd.concat([india_weather,us_weather],axis =1)
df
```

```python
# now we will merge the datafreames
temperature_df = pd.DataFrame({'city': ['mumbai','delhi','banglore','hyderabad'],'temperature':[34,45,43,23]})
temperature_df
```

```python
humidity_df =pd.DataFrame({'city':['mumbai','delhi','banglore'],'humidity':[32,54,54]})
humidity_df
```

```python
#Mergine two dataframes
df= pd.merge(temperature_df,humidity_df)
df #while merging hyderabad is not showing because humidity element is not mentioned
```

```python
df =pd.merge(temperature_df,humidity_df,on ='city',how = 'outer')
df
```
